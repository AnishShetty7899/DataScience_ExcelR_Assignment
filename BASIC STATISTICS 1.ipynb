### outliers and bound calcuation

```python
#outliers calculation
print("For Volumn column:")
q1 = np.quantile(df['Volume'], 0.25)
q2 = np.quantile(df['Volume'], 0.50)
q3 = np.quantile(df['Volume'], 0.75)
iqr = q3-q1

# finding upper and lower bonds
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print("Interquantile Range:[",q3,",",q1,"]")
print('The following are the outliers in the boxplot:')
for i in df['Volume']:
    if((i <= lower_bound) | (i >= upper_bound)):
        print(i, end=",")
```
